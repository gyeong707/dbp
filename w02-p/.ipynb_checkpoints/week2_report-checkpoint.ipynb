### 1. 새로 배운 내용

강의에 나온 php, mysql 전부 다 처음 사용해보는 것이었기 때문에 배워가는 내용이 정말 많았습니다.  
처음에 클라이언트, 서버, php, 그리고 데이터베이스가 어떻게 서로 교류하면서 작동하는지 알려주신 것이 학습 진행에 있어 큰 도움이 되었다고 생각합니다. 이 부분에 대해 내용을 정리해보자면 다음과 같습니다. 

#### 1-1. 환경에 대한 이해
- 클라이언트 : 원하는 정보를 네트워크를 통해 서버에게 요청함.
- 서버 : 요청된 정보를 데이터베이스로부터 찾아서 클라이언트에게 전달
- 데이터베이스 : 모든 정보가 저장되어 있는 곳
- PHP : 동적 페이지 관리의 역할.   
클라이언트는 웹 브라우저라고 할수 있으며, 웹 브라우저는 서버로부터 전달받은 정보를 화면에 띄운다.  
전체적인 구조는  
클라이언트 - 서버 - PHP - 데이터베이스로 정보를 요청하고 반환받는 형태이다.
<br>

#### 1-2. PHP에 대해 조금 더 살펴보자.
사용자가 클라이언트에게 특정 정보를 요청할 때, 웹 서버는 요청 정보를 확인한다.   
그런데 그 안에 php 코드가 있다? 웹 서버는 html 코드만 읽을 수 있기 때문에 이 php 관련 코드 내용은 php로 전달을 한다.   
그러면 php는 그 코드를 해석하고, 어떤 정보를 요청하는지 확인하며, 해당 정보를 데이터베이스로에게 요청한다.   
데이터베이스가 요청에 맞는 정보를 반환하면 php는 해당 정보를 가지고 html 코드를 쭉 작성하고, 최종적으로 웹 서버가 이해할 수 있는 태그 형태로 이루어진 페이지를 만들게 된다. 

#### 1-3. 공식문서를 읽는 방법  
공식 문서를 보면서 설명을 해주셨기 때문에, 사용 에제, 반환값, input 파라미터 등을 보여주셨던 것이 좋았습니다. 잘 보고 배워서 나중에 혼자서도 공식문서를 보고 필요한 기능을 찾아 사용할 수 있으면 좋겠습니다.

### 2. 문제 발생 / 고민 / 해결과정

#### 2-1. 비밀번호 문제
처음 헤맸던 점은, 기억력이 나쁜 편이라 1주차 교수님 강의를 통해 
교수님과 똑같이 id, password를 지정하였는데, 이게 웬열, 오늘 로그인을 하려고보니 자꾸 에러메시지가 발생했다.
구글링을 통해 이건 비밀번호가 맞지 않아  발생하는 문제라는 뜻을 보고 '그럴리가 없는데...'하다가 
혹시 몰라서 capslock을 켜 대문자로 써보았더니 됐다. 
잘 기억하자. 내 비밀번호는 'rootroot'가 아니라 'ROOTROOT'이다. 

#### 2-2. 테이블 초기화
과제하면서 새로운 데이터베이스를 만들었는데,
일부 데이터가 입력되지 않은 채로 행이 만들어졌다. 이를 위해 나는 테이블을 초기화하는 명령어를 찾았고
TRUNCATE 'TABLENAME'; 명령어를 통해 쉽게 테이블을 초기화할 수 있었다.
이건 초반이라 그랬지, 만약에 중간에 테이블에 행을 잘못 넣었다면,
일부 행만 삭제할 수 있는 새로운 명령어를 공부해야하겠다라는 생각이 들었다.


#### 2-3. undefined index error
Notice: Undefined index: name in C:\Bitnami\wampstack-7.4.9-0\apache2\htdocs\w02-p\process_create.php on line 7
자꾸 create.php에서 보낸 정보를 입력한는 process_create.php에서 오류가 발생했다.
name이 입력이 안되는 문제가 자꾸만 일어난다. 
undefined index라니! 무슨 문제일까?  
<해결방안>   딱히 어떤것이 문제인지는 정말 잘 모르겠다. 내가 post를 보내는 것이 name으로 가지 않는 것일까 하여, 코드에 
'var_dump($_POST)' 를 추가했더니, 얼레'name'으로 잘 들어오잖아? 하는 순간 모든 것이 해결되었다.
도대체 무엇이 문제였을까? 찝찝하긴 하지만, 해결은 되어서 다행이다.
